```python
import numpy as np
import gym
import time
import matplotlib.pyplot as plt
import cv2
```

```python
#Make the environment
env = gym.make("DemonAttack-v0")
```

```python
env.unwrapped.get_action_meanings() 
```

```python
obs = env.reset()
```

```python
def downsample_image(A):
    B = A[31:195]
    B = B.mean(axis =2)
    B = B/255.0
    # B = cv2.resize(A, (80, 80))
    B = B[20:155, :]
    return B
```

```python
plt.imshow(obs)
```

```python
obs_t = downsample_image(obs)
plt.imshow(obs_t)
```

```python
for i in range(10):
    obs, reward, done,_ = env.step(1)
    obs = downsample_image(obs)
    plt.figure()
    plt.imshow(obs)
    plt.show()
```
